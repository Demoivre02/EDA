```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime


df = pd.read_csv(r'10k_synthea_covid19_csv/patients.csv')

def calculate_age(birthdate):

    born = datetime.strptime(birthdate, '%Y-%m-%d')
    today = datetime.now()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

def analyze_age_distribution(dataframe):
   
    dataframe['AGE'] = dataframe['BIRTHDATE'].apply(calculate_age)
    

    age_stats = {
        'mean_age': dataframe['AGE'].mean(),
        'median_age': dataframe['AGE'].median(),
        'min_age': dataframe['AGE'].min(),
        'max_age': dataframe['AGE'].max()
    }
    
    plt.figure(figsize=(10, 6))
    plt.hist(dataframe['AGE'], bins=20, edgecolor='black')
    plt.title('Age Distribution of Patients')
    plt.xlabel('Age')
    plt.ylabel('Number of Patients')
    plt.close()
    
    return age_stats

age_stats = analyze_age_distribution(df)

print("Age Distribution Statistics:")
for stat, value in age_stats.items():
    print(f"{stat.replace('_', ' ').title()}: {value:.2f}")


print("\nFirst few rows with calculated ages:")
print(df[['BIRTHDATE', 'AGE']].head())

age_groups = pd.cut(df['AGE'], bins=[0, 18, 30, 45, 60, 75, 100], labels=['0-18', '19-30', '31-45', '46-60', '61-75', '76+'])
age_group_percentages = age_groups.value_counts(normalize=True) * 100

print("\nAge Group Percentages:")
print(age_group_percentages)

plt.figure(figsize=(10, 8))
plt.pie(age_group_percentages, labels=age_group_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title('Age Group Distribution of Patients')
plt.axis('equal')
plt.close()
```

```python
import pandas as pd
from collections import Counter
from datetime import datetime


patients_df = pd.read_csv(r'10k_synthea_covid19_csv/patients.csv')


conditions_list = []
with open('10k_synthea_covid19_csv/conditions.txt', 'r') as file:
    for line in file:
        line = line.strip()
        if line.startswith("'") and line.endswith("',"):
            conditions_list.append(line[1:-2])  

conditions_df = pd.DataFrame({'Condition': conditions_list})

condition_counts = Counter(conditions_list)

top_10_conditions = dict(condition_counts.most_common(10))

print("Top 10 most common conditions:")
for condition, count in top_10_conditions.items():
    print(f"{condition}: {count}")

print("\nDemographic analysis:")
print(f"Total patients: {len(patients_df)}")
print(f"Gender distribution:\n{patients_df['GENDER'].value_counts(normalize=True)}")
print(f"\nRace distribution:\n{patients_df['RACE'].value_counts(normalize=True)}")
print(f"\nEthnicity distribution:\n{patients_df['ETHNICITY'].value_counts(normalize=True)}")

patients_df['BIRTHDATE'] = pd.to_datetime(patients_df['BIRTHDATE'])
current_date = datetime.now()
patients_df['AGE'] = current_date.year - patients_df['BIRTHDATE'].dt.year

patients_df.loc[
    (current_date.month < patients_df['BIRTHDATE'].dt.month) | 
    ((current_date.month == patients_df['BIRTHDATE'].dt.month) & (current_date.day < patients_df['BIRTHDATE'].dt.day)),
    'AGE'
] -= 1

print(f"\nAverage age: {patients_df['AGE'].mean():.2f} years")

print("\nTop 5 conditions by gender (count):")
for gender in patients_df['GENDER'].unique():
    gender_conditions = [cond for cond, count in condition_counts.items() if count > 0]
    print(f"\n{gender}:")
    for cond in sorted(gender_conditions, key=lambda x: condition_counts[x], reverse=True)[:5]:
        print(f"{cond}: {condition_counts[cond]}")

patients_df['AGE_GROUP'] = pd.cut(patients_df['AGE'], bins=[0, 18, 30, 50, 70, 100], labels=['0-18', '19-30', '31-50', '51-70', '71+'])
print("\nTop 5 conditions by age group (count):")
for age_group in patients_df['AGE_GROUP'].unique():
    age_group_conditions = [cond for cond, count in condition_counts.items() if count > 0]
    print(f"\n{age_group}:")
    for cond in sorted(age_group_conditions, key=lambda x: condition_counts[x], reverse=True)[:5]:
        print(f"{cond}: {condition_counts[cond]}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from datetime import datetime

def is_jupyter():
    try:
        shell = get_ipython().__class__.__name__
        if shell == 'ZMQInteractiveShell':
            return True   
        elif shell == 'TerminalInteractiveShell':
            return False 
        else:
            return False 
    except NameError:
        return False     


if is_jupyter():
    %matplotlib inline
else:
    plt.switch_backend('Agg')


patients_df = pd.read_csv(r'10k_synthea_covid19_csv/patients.csv')


conditions_list = []
with open('10k_synthea_covid19_csv/conditions.txt', 'r') as file:
    for line in file:
        line = line.strip()
        if line.startswith("'") and line.endswith("',"):
            conditions_list.append(line[1:-2])  


patients_df['BIRTHDATE'] = pd.to_datetime(patients_df['BIRTHDATE'])
current_date = datetime.now()
patients_df['AGE'] = current_date.year - patients_df['BIRTHDATE'].dt.year
patients_df.loc[
    (current_date.month < patients_df['BIRTHDATE'].dt.month) | 
    ((current_date.month == patients_df['BIRTHDATE'].dt.month) & (current_date.day < patients_df['BIRTHDATE'].dt.day)),
    'AGE'
] -= 1


plt.figure(figsize=(10, 6))
plt.hist(patients_df['AGE'], bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
if not is_jupyter():
    plt.savefig('/tmp/outputs/age_distribution.png')
    plt.close()
else:
    plt.show()


top_5_conditions = [cond for cond, _ in Counter(conditions_list).most_common(5)]
gender_condition_counts = {gender: {cond: 0 for cond in top_5_conditions} for gender in patients_df['GENDER'].unique()}

for gender in patients_df['GENDER'].unique():
    gender_conditions = [cond for cond, g in zip(conditions_list, patients_df['GENDER'].repeat(len(conditions_list) // len(patients_df))) if g == gender]
    for cond in top_5_conditions:
        gender_condition_counts[gender][cond] = gender_conditions.count(cond)

plt.figure(figsize=(12, 6))
bottom = np.zeros(len(gender_condition_counts))
for condition in top_5_conditions:
    values = [gender_condition_counts[gender][condition] for gender in gender_condition_counts]
    plt.bar(list(gender_condition_counts.keys()), values, bottom=bottom, label=condition)
    bottom += values

plt.title('Top 5 Conditions by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
if not is_jupyter():
    plt.savefig('/tmp/outputs/conditions_by_gender.png')
    plt.close()
else:
    plt.show()

print("Script execution completed. If not in Jupyter, check the '/tmp/outputs' folder for the generated plots.")
```

```python
# This cell contains the necessary imports and magic commands
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from datetime import datetime

# Read the patients data
patients_df = pd.read_csv(r'10k_synthea_covid19_csv/patients.csv')

# Read the conditions data
conditions_list = []
with open('10k_synthea_covid19_csv/conditions.txt', 'r') as file:
    for line in file:
        line = line.strip()
        if line.startswith("'") and line.endswith("',"):
            conditions_list.append(line[1:-2])  # Remove quotes and comma

# Calculate age
patients_df['BIRTHDATE'] = pd.to_datetime(patients_df['BIRTHDATE'])
current_date = datetime.now()
patients_df['AGE'] = current_date.year - patients_df['BIRTHDATE'].dt.year
patients_df.loc[
    (current_date.month < patients_df['BIRTHDATE'].dt.month) | 
    ((current_date.month == patients_df['BIRTHDATE'].dt.month) & (current_date.day < patients_df['BIRTHDATE'].dt.day)),
    'AGE'
] -= 1

# 1. Age distribution histogram
plt.figure(figsize=(10, 6))
plt.hist(patients_df['AGE'], bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('/tmp/outputs/age_distribution.png')
plt.close()

# 2. Birth year trends
plt.figure(figsize=(12, 6))
patients_df['BIRTH_YEAR'] = patients_df['BIRTHDATE'].dt.year
birth_year_counts = patients_df['BIRTH_YEAR'].value_counts().sort_index()
plt.plot(birth_year_counts.index, birth_year_counts.values)
plt.title('Number of Patients Born per Year')
plt.xlabel('Birth Year')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.savefig('/tmp/outputs/birth_year_trends.png')
plt.close()

# 3. Conditions by age group
age_bins = [0, 18, 30, 45, 60, 75, 100]
age_labels = ['0-18', '19-30', '31-45', '46-60', '61-75', '76+']
patients_df['AGE_GROUP'] = pd.cut(patients_df['AGE'], bins=age_bins, labels=age_labels, right=False)

top_5_conditions = [cond for cond, _ in Counter(conditions_list).most_common(5)]
age_condition_counts = {age_group: {cond: 0 for cond in top_5_conditions} for age_group in age_labels}

for age_group in age_labels:
    age_group_conditions = [cond for cond, a in zip(conditions_list, patients_df['AGE_GROUP'].repeat(len(conditions_list) // len(patients_df))) if a == age_group]
    for cond in top_5_conditions:
        age_condition_counts[age_group][cond] = age_group_conditions.count(cond)

plt.figure(figsize=(12, 6))
bottom = np.zeros(len(age_condition_counts))
for condition in top_5_conditions:
    values = [age_condition_counts[age_group][condition] for age_group in age_condition_counts]
    plt.bar(list(age_condition_counts.keys()), values, bottom=bottom, label=condition)
    bottom += values

plt.title('Top 5 Conditions by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('/tmp/outputs/conditions_by_age_group.png')
plt.close()

print("Analysis completed. Please check the '/tmp/outputs/' directory for the generated plots.")
```
