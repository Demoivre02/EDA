```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv(r'10k_synthea_covid19_csv/patients.csv')

def count_patients(dataframe):
    """
    Function to determine the number of patients in the dataset
    
    Args:
    dataframe (pd.DataFrame): The dataframe containing patient data
    
    Returns:
    int: The number of patients in the dataset
    """
    return len(dataframe)

# Count the number of patients
num_patients = count_patients(df)

print(f"The number of patients in the dataset is: {num_patients}")

# Display the first few rows of the dataframe to get an idea of the data
print("\nFirst few rows of the dataset:")
print(df.head())

# Display basic information about the dataframe
print("\nDataframe info:")
df.info()

# Display summary statistics of the numerical columns
print("\nSummary statistics:")
print(df.describe())
```
